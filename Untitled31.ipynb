```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
data = pd.read_csv(next(iter(uploaded.keys())))
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
import shap

sns.set(style="whitegrid")
```

```python
data['Amount_scaled'] = StandardScaler().fit_transform(data[['Amount']])
data.drop(['Time','Amount'], axis=1, inplace=True)
X = data.drop('Class', axis=1)
y = data['Class']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, stratify=y, random_state=42)
X_train_res, y_train_res = SMOTE(random_state=42).fit_resample(X_train, y_train)
```

```python
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(X_train_res, y_train_res)
y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]
```

```python
print(classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```

```python
# Only compute for 100 samples
sample_data = X_test.iloc[:100]

# Compute SHAP values
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(sample_data)

# Use the same subset for plotting
shap.summary_plot(shap_values, sample_data)
```

```python
plt.figure()
shap.summary_plot(shap_values, X_test, plot_type="bar", show=False)
plt.show()
```

```python
plt.figure()
shap.summary_plot(shap_values, sample_data, show=False)
plt.show()
```

```python
sample_idx = 10
shap.initjs()
display(shap.force_plot(
    explainer.expected_value, shap_values[sample_idx], X_test.iloc[sample_idx]))
```

```python
# Use only the subset
sample_data = X_test.iloc[:100]

# Top feature
top_feat = sample_data.columns[np.abs(shap_values).mean(axis=0).argmax()]

# Dependence plot
shap.dependence_plot(top_feat, shap_values, sample_data, show=False)
plt.show()
```

```python
# Check if shap_values is a list (multi-class) or array (binary)
if isinstance(shap_values, list):
    shap_values_class1 = shap_values[1]  # For class 1
else:
    shap_values_class1 = shap_values

print("SHAP shape:", shap_values_class1.shape)

# Check feature count
if shap_values_class1.shape[1] == 1:
    print("❌ Only 1 feature detected in SHAP. Dependence plot needs multiple features.")
else:
    # Plot
    top_feat = sample_data.columns[np.abs(shap_values_class1).mean(axis=0).argmax()]
    shap.dependence_plot(top_feat, shap_values_class1, sample_data, show=False)
    plt.show()

```
